# 함수
> #### <span style="color:red">부호함수</span> <br> 부호함수(sign function)는 입력이 양수이면 1, 음수이면 -1, 0이면 0을 출력하는 x=0 에서 불연속인 함수 <br>


> #### <span style="color:red">단위 계단함수</span> <br> 단위계단함수(Heaviside step function)도 x=0에서 불연속인 함수이다. 넘파이 구현이 없으므로 직접 구현해야 한다.<br>





```python
def heaviside_step(x):
    if isinstance(x, np.ndarray):
        return np.where(x >= 0, 1, 0)
    else:
        return 1.0 if x >= 0 else 0.0 

heaviside_step(-0.0001), heaviside_step(0), heaviside_step(0.0001)    
```

# <span style="color:red"> 함수의 그래프 </span>
> ##### 함수의 형상을 시각적으로 파악하기 위해 그래프 또는 플롯을 사용하기도 한다.

> #### 독립변숫값 <span style="color:red">벡터 x</span><br> ->이 위치에 대한 함숫값 <span style="color:red">벡터 y</span>를 하나하나 계산하여 <br>이렇게 만들어진 점을 직선으로 연결하는 방식으로 플롯을 그린다

$$ f(x)=x^3−3x^2+x   그래프

```python
def f(x):
    return x**3 - 3 * x**2 + x


# x=-1부터 3까지 9개점을 찍는다 (0.5간격으로 찍히겠죠?)
# x=np.linspace(-1,3,9)

# x=-1부터 3까지 400개점을 찍는다 (촘촘한 간격으로 찍히겠죠?)
x=np.linspace(-1,3,400)

x

y=f(x)
y


# 그래프 그리기
plt.plot(x, y, '-gx')  # green x 표
plt.xlim(-2, 4)  # x 축 범위 설정
plt.title("함수 $f(x) = x^3 - 3x^2 + x$의 그래프")
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(np.arange(-1, 4))  # x 축 눈금 설정
plt.yticks(np.arange(-5, 4))  # y 축 눈금 설정
plt.grid(True)  # 격자 추가
plt.show()
```

## 역함수 그래프

> 역함수의 그래프는 원래의 함수에서 x 축과 y축이 바뀐것으로 y=x가 나타내는 직선(원점을 통과하는 기울기 1인 직선)을 대칭축으로 대칭인 함수의 그래프와 같다

```python
def f1(x):
    return x ** 2


def f1inv(x):
    return np.sqrt(x)


x = np.linspace(0, 3, 300)
plt.plot(x, f1(x), "r-", label="함수 $f(x) = x^2$")
plt.plot(x, f1inv(x), "b-.", label="역함수 $f^{-1}(x) = \sqrt{x}$")
plt.plot(x, x, "g--")
plt.axis("equal")

plt.xlim(0, 2)
plt.ylim(0, 2)
plt.legend()
plt.title("역함수의 그래프")
plt.show()
```

> #### <span style="color:red">3. 지수함수</span> 정해진 어떤 값을 입력값으로 거듭제곱하여 출력을 구하는 함수
$$ y=a^x $$
> 거듭제곱을  할때 아래있는 수를 밑이라함
> 밑을 오일러수 e(약 2.718)로 하여 거듭제곱을 하는 함수를 지수함수라고 함
$$ y= e^x $$
지수함수는 다음처럼 표시할수도 있음
$$ y = \exp(x) = \exp{x} $$




```python
np.exp(-10), np.exp(-1), np.exp(-0.1), np.exp(0), np.exp(0.1), np.exp(1), np.exp(10)
```

```python
## 지수함수의 특성
# 1. 양수(e)를 거듭 제곱한 값이므로 항상 양수
# 2. x=0일때 1이 됌
# 3. x가 양의 무한대로 가면 (x→∞), 양의 무한대로 다가감
# 4. x가 음의 무한대로 가면(x→−∞), 0으로 다가간다
# 5. x1>x2이면 exp(x1)> exp(x2)이다.
xx = np.linspace(-2, 2, 100)
yy = np.exp(xx)
plt.title("지수함수")
plt.plot(xx, yy)
plt.axhline(1, c='r', ls="--")
plt.axhline(0, c='r', ls="--")
plt.axvline(0, c='r', ls="--")
plt.xlabel("$x$")
plt.ylabel("$\exp(x)$")
plt.show()
```

f(x_1) \cdot f(x_2) = e^{x_1} \cdot e^{x_2} = e^{x_1 + x_2} = f(x_1 + x_2)
$$

> #### <span style="color:red">4. 로지스틱함수</span> 지수함수를 변형한 함수로 회귀 분석이나 인공신경망에서 자주 사옹됌(시그모이드 함수 종류의 하나)
> 시그모이드 함수중에서 로지스틱함수가 가장 널리 쓰이기 때문에 보통 시그모이드 함수라 하면 로지스틱함수를 의미
$$ σ(x)= \frac{1}{1+exp(-x)} $$


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    return np.exp(-x)

# Create an array of x values from -10 to 10
x = np.linspace(-10, 10, 400)

# Compute the corresponding y values
y = f(x)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$f(x) = \exp(-x)$')
plt.title('Graph of $f(x) = \exp(-x)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```

```python
# 로지스틱(시그모이드) 함수
def logistic(x):
    return 1 / (1 + np.exp(-x)) # x가 커질수록 분모가 작아짐 -> 결국 수자체는 커짐


# 로지스틱(시그모이드) 함수
def exp(x):
    return  np.exp(-x) # x가 커질수록 분모가 작아짐 -> 결국 수자체는 커짐

xx = np.linspace(-10,10, 100)
print(exp(xx)) # x가 커질수록 y값은 작아짐 왜냐면 exp(-x)=> -거듭제곱이기 때문
print("================")
print(logistic(xx))
plt.plot(xx, logistic(xx))
plt.title("로지스틱함수")
plt.xlabel("$x$")
plt.ylabel("$\sigma(x)$")
plt.show()
```

> #### <span style="color:red">5. 로그함수:지수함수와 역함수 관계</span> 
> e(어떤수)를 거듭제곱하여 특정한 수 a가 되도록 하는 수를 log a라 표기하고 log라 읽음

```python
print( np.exp(2.3025851)) # 10
print(np.log(10)) # 2.302585
```

$$ y= a^x  = e^{(loga)x} = e^{x loga}

```python
# 로그함수는 다음과 같은 특징을 가진다
# 1. x값, 즉 입력변수값이 양수여야 함. 0이거나 음수면 정의되지 않음
# 2. x>1면 y>0(양수)
# 3. x=1이면 y=0
# 4. 0<x<1 면 y<0(음수)
# 5.x1>x2면 logx1>logx2이다
xx = np.linspace(0.01, 8, 100)
yy = np.log(xx)
plt.title("로그함수")
plt.plot(xx, yy)
plt.axhline(0, c='r', ls="--")
plt.axvline(0, c='r', ls="--")
plt.axvline(1, c='r', ls="--")
plt.xlabel("$x$")
plt.ylabel("$\log(x)$")
plt.show()
```
