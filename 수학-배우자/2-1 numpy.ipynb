<span style='color:yellow'> 
벡터의 차원: 원소의 개수 <br></span>
 > n차원 벡터: n개의 요소를 갖는 벡터, 예: [1, 2, 3]은 3차원 벡터<br><br>

<span style='color:yellow'> 
배열의 차원:  줄수 (원소의 개수가 몇개이든 한줄로 나타낼 수 있다면 1차원 배열)<br></span>
> 1차원 배열: 단일 행 또는 열로 구성된 배열, 예: [1, 2, 3]
 2차원 배열: 행과 열을 포함한 테이블 형태의 배열, 예: [[1, 2, 3], [4, 5, 6]]
 3차원 배열: 여러 개의 2차원 배열을 층(layer)로 쌓은 형태, 예를 들어 이미지 데이터에서 색상 채널을 표현할 때 사용됨




```python
# 이 데이터는 0부터 9까지의 숫자를 손으로 쓴 후에 8x8 해상도의 이미지로 스캔한 것이다. 
# 다음 코드는 그중 숫자 0과 1의 이미지 8개를 출력한다.

from sklearn.datasets import load_digits  # 손글씨 숫자 (0~9) 데이터 세트
import matplotlib.pyplot as plt

# data: 각 이미지가 평탄화(flattened된 형태로 나타난 2차원 배열)
#    -> 8x8 픽셀 크기의 이미지를 일렬로 펼친 64개의 픽셀 값으로 구성되어 있음 (0~16까지의 그레이스케일 값을 가짐)

# target: 각 이미지에 대응되는 실제 숫자 레이블 (이 배열은 각 이미지가 어떤 숫자를 나타내는지를 정수 형태로 포함)
# frame: 데이터 프레임 형식으로 데이터를 반환할때 사용되는 옵션
# feature_names: 각 픽셀 위치의 이름-> pixel_0_0은 첫번째 행의 첫번째 열 픽셀 , pixel_0_1 은 첫행의 두번째 열
# target_names: 대상 레이블에 해당하는 숫자 이름
# images: 원본 8x8 픽셀 이미지 데이터 -> data 필드의 정보를 2차원 형태로 복원한 것

# images는 그림이고 data는 머신러닝에 적합하게 images를 변형
# images의 첫번째 요소가 2차원 배열 [ [0, 0, 5, ...], [0, 0, 13, ...], ...]
# data는 평탄화(1차원 배열로)  [0, 0, 5, ..., 0, 0, 13, ...]


# 0부터 9까지 손글씨 데이터

digits=load_digits() # 데이터 로드

print(digits)
sample =[0,10,20,30,1,11,21,31] # 선택된 이미지 번호
d=[]
for i in range(8):
    d.append(digits.images[sample[i]])

plt.figure(figsize=(8,2)) # 그림의 크기를 8x2 인치로 설정

# i-> 0부터 7까지
for i in range(8):
    plt.subplot(1,8,i+1) # subplot(행의수,열의수,인덱스번호)-> 그림내 여러개의 서브플롯을 배열형식으로 배치 
    # imshow: 배열의 형태로 제공된 이미지 데이터를 시각화 하는 함수
    #  >  'nearest'는 가장 가까운 픽셀의 값을 그대로 사용하여 확대하는 방법
    #  > cmap=plt.cm.bone_r : 컬러맵 지정 bone_r로 배경은 흰색 , 숫자들이 어두운 배경으로 나타남
    plt.imshow(d[i],interpolation='nearest',cmap=plt.cm.bone_r) 
    plt.grid(False)

    #  각각 x축과 y축의 눈금을 설정 -> 빈리스트로 시각화 방해되는 요소 최소화
    plt.xticks([])
    plt.yticks([])
    plt.title("iamge {}".format(i+1))
plt.suptitle("숫자 0과 1이미지")
plt.tight_layout()
plt.show()
```

#### <span style="color:red">이 2차원 이미지를 64크기의 1차원 벡터로 펼치면 다음과 같다</span>

###### > 같은 숫자에 대한 벡터가 서로 닮음

```python
v = []
for i in range(8):
    v.append(d[i].reshape(64, 1))  # 벡터화-> digits.images에서 각 이미지가 8x8 픽셀의 크기를 가짐
# 이미지 8x8을 각각 단일열로 나타내기 위해-> 64x1 크기의 벡터로 변환한것


# 벡터화는 머신러닝에서 이미지 처리할때 필수적인 단계-> 일관되고 효율적으로 처리 가능
 
plt.figure(figsize=(8, 3))
for i in range(8):
    plt.subplot(1, 8, i + 1)

    # v[i]: 여기서 v[i]는 벡터화된 이미지 데이터의 i번째 아이템
    # aspect: 막대기 가로길이
    # cmap=plt.cm.bone_r: 맵 색상
    plt.imshow(v[i], aspect=0.4,
               interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False); plt.xticks([]); plt.yticks([])
    plt.title("vector {}".format(i + 1))
plt.suptitle("vetorized images", y=1.05)
plt.tight_layout(w_pad=7)
plt.show()
```

> 이를 한줄의 긴 벡터로 변환하여 각 픽셀의 밝기(강도)가  벡터의 각 요소에 저장됌
