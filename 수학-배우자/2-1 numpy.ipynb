```python
# Ïù¥ Îç∞Ïù¥ÌÑ∞Îäî 0Î∂ÄÌÑ∞ 9ÍπåÏßÄÏùò Ïà´ÏûêÎ•º ÏÜêÏúºÎ°ú Ïì¥ ÌõÑÏóê 8x8 Ìï¥ÏÉÅÎèÑÏùò Ïù¥ÎØ∏ÏßÄÎ°ú Ïä§Ï∫îÌïú Í≤ÉÏù¥Îã§. 
# Îã§Ïùå ÏΩîÎìúÎäî Í∑∏Ï§ë Ïà´Ïûê 0Í≥º 1Ïùò Ïù¥ÎØ∏ÏßÄ 8Í∞úÎ•º Ï∂úÎ†•ÌïúÎã§.

from sklearn.datasets import load_digits  # ÏÜêÍ∏ÄÏî® Ïà´Ïûê (0~9) Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏
import matplotlib.pyplot as plt

# data: Í∞Å Ïù¥ÎØ∏ÏßÄÍ∞Ä ÌèâÌÉÑÌôî(flattenedÎêú ÌòïÌÉúÎ°ú ÎÇòÌÉÄÎÇú 2Ï∞®Ïõê Î∞∞Ïó¥)
#    -> 8x8 ÌîΩÏÖÄ ÌÅ¨Í∏∞Ïùò Ïù¥ÎØ∏ÏßÄÎ•º ÏùºÎ†¨Î°ú ÌéºÏπú 64Í∞úÏùò ÌîΩÏÖÄ Í∞íÏúºÎ°ú Íµ¨ÏÑ±ÎêòÏñ¥ ÏûàÏùå (0~16ÍπåÏßÄÏùò Í∑∏Î†àÏù¥Ïä§ÏºÄÏùº Í∞íÏùÑ Í∞ÄÏßê)

# target: Í∞Å Ïù¥ÎØ∏ÏßÄÏóê ÎåÄÏùëÎêòÎäî Ïã§Ï†ú Ïà´Ïûê Î†àÏù¥Î∏î (Ïù¥ Î∞∞Ïó¥ÏùÄ Í∞Å Ïù¥ÎØ∏ÏßÄÍ∞Ä Ïñ¥Îñ§ Ïà´ÏûêÎ•º ÎÇòÌÉÄÎÇ¥ÎäîÏßÄÎ•º Ï†ïÏàò ÌòïÌÉúÎ°ú Ìè¨Ìï®)
# frame: Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ ÌòïÏãùÏúºÎ°ú Îç∞Ïù¥ÌÑ∞Î•º Î∞òÌôòÌï†Îïå ÏÇ¨Ïö©ÎêòÎäî ÏòµÏÖò
# feature_names: Í∞Å ÌîΩÏÖÄ ÏúÑÏπòÏùò Ïù¥Î¶Ñ-> pixel_0_0ÏùÄ Ï≤´Î≤àÏß∏ ÌñâÏùò Ï≤´Î≤àÏß∏ Ïó¥ ÌîΩÏÖÄ , pixel_0_1 ÏùÄ Ï≤´ÌñâÏùò ÎëêÎ≤àÏß∏ Ïó¥
# target_names: ÎåÄÏÉÅ Î†àÏù¥Î∏îÏóê Ìï¥ÎãπÌïòÎäî Ïà´Ïûê Ïù¥Î¶Ñ
# images: ÏõêÎ≥∏ 8x8 ÌîΩÏÖÄ Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞ -> data ÌïÑÎìúÏùò Ï†ïÎ≥¥Î•º 2Ï∞®Ïõê ÌòïÌÉúÎ°ú Î≥µÏõêÌïú Í≤É

# imagesÎäî Í∑∏Î¶ºÏù¥Í≥† dataÎäî Î®∏Ïã†Îü¨ÎãùÏóê Ï†ÅÌï©ÌïòÍ≤å imagesÎ•º Î≥ÄÌòï
# imagesÏùò Ï≤´Î≤àÏß∏ ÏöîÏÜåÍ∞Ä 2Ï∞®Ïõê Î∞∞Ïó¥ [ [0, 0, 5, ...], [0, 0, 13, ...], ...]
# dataÎäî ÌèâÌÉÑÌôî(1Ï∞®Ïõê Î∞∞Ïó¥Î°ú)  [0, 0, 5, ..., 0, 0, 13, ...]


# 0Î∂ÄÌÑ∞ 9ÍπåÏßÄ ÏÜêÍ∏ÄÏî® Îç∞Ïù¥ÌÑ∞

digits=load_digits() # Îç∞Ïù¥ÌÑ∞ Î°úÎìú

print(digits)
sample =[0,10,20,30,1,11,21,31] # ÏÑ†ÌÉùÎêú Ïù¥ÎØ∏ÏßÄ Î≤àÌò∏
d=[]
for i in range(8):
    d.append(digits.images[sample[i]])

plt.figure(figsize=(8,2)) # Í∑∏Î¶ºÏùò ÌÅ¨Í∏∞Î•º 8x2 Ïù∏ÏπòÎ°ú ÏÑ§Ï†ï

# i-> 0Î∂ÄÌÑ∞ 7ÍπåÏßÄ
for i in range(8):
    plt.subplot(1,8,i+1) # subplot(ÌñâÏùòÏàò,Ïó¥ÏùòÏàò,Ïù∏Îç±Ïä§Î≤àÌò∏)-> Í∑∏Î¶ºÎÇ¥ Ïó¨Îü¨Í∞úÏùò ÏÑúÎ∏åÌîåÎ°ØÏùÑ Î∞∞Ïó¥ÌòïÏãùÏúºÎ°ú Î∞∞Ïπò 
    # imshow: Î∞∞Ïó¥Ïùò ÌòïÌÉúÎ°ú Ï†úÍ≥µÎêú Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞Î•º ÏãúÍ∞ÅÌôî ÌïòÎäî Ìï®Ïàò
    #  >  'nearest'Îäî Í∞ÄÏû• Í∞ÄÍπåÏö¥ ÌîΩÏÖÄÏùò Í∞íÏùÑ Í∑∏ÎåÄÎ°ú ÏÇ¨Ïö©ÌïòÏó¨ ÌôïÎåÄÌïòÎäî Î∞©Î≤ï
    #  > cmap=plt.cm.bone_r : Ïª¨Îü¨Îßµ ÏßÄÏ†ï bone_rÎ°ú Î∞∞Í≤ΩÏùÄ Ìù∞ÏÉâ , Ïà´ÏûêÎì§Ïù¥ Ïñ¥ÎëêÏö¥ Î∞∞Í≤ΩÏúºÎ°ú ÎÇòÌÉÄÎÇ®
    plt.imshow(d[i],interpolation='nearest',cmap=plt.cm.bone_r) 
    plt.grid(False)

    #  Í∞ÅÍ∞Å xÏ∂ïÍ≥º yÏ∂ïÏùò ÎààÍ∏àÏùÑ ÏÑ§Ï†ï -> ÎπàÎ¶¨Ïä§Ìä∏Î°ú ÏãúÍ∞ÅÌôî Î∞©Ìï¥ÎêòÎäî ÏöîÏÜå ÏµúÏÜåÌôî
    plt.xticks([])
    plt.yticks([])
    plt.title("iamge {}".format(i+1))
plt.suptitle("Ïà´Ïûê 0Í≥º 1Ïù¥ÎØ∏ÏßÄ")
plt.tight_layout()
plt.show()
```

#### <span style="color:red">Ïù¥ 2Ï∞®Ïõê Ïù¥ÎØ∏ÏßÄÎ•º 64ÌÅ¨Í∏∞Ïùò 1Ï∞®Ïõê Î≤°ÌÑ∞Î°ú ÌéºÏπòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§</span>

###### > Í∞ôÏùÄ Ïà´ÏûêÏóê ÎåÄÌïú Î≤°ÌÑ∞Í∞Ä ÏÑúÎ°ú ÎãÆÏùå

```python
v = []
for i in range(8):
    v.append(d[i].reshape(64, 1))  # Î≤°ÌÑ∞Ìôî-> digits.imagesÏóêÏÑú Í∞Å Ïù¥ÎØ∏ÏßÄÍ∞Ä 8x8 ÌîΩÏÖÄÏùò ÌÅ¨Í∏∞Î•º Í∞ÄÏßê
# Ïù¥ÎØ∏ÏßÄ 8x8ÏùÑ Í∞ÅÍ∞Å Îã®ÏùºÏó¥Î°ú ÎÇòÌÉÄÎÇ¥Í∏∞ ÏúÑÌï¥-> 64x1 ÌÅ¨Í∏∞Ïùò Î≤°ÌÑ∞Î°ú Î≥ÄÌôòÌïúÍ≤É


# Î≤°ÌÑ∞ÌôîÎäî Î®∏Ïã†Îü¨ÎãùÏóêÏÑú Ïù¥ÎØ∏ÏßÄ Ï≤òÎ¶¨Ìï†Îïå ÌïÑÏàòÏ†ÅÏù∏ Îã®Í≥Ñ-> ÏùºÍ¥ÄÎêòÍ≥† Ìö®Ïú®Ï†ÅÏúºÎ°ú Ï≤òÎ¶¨ Í∞ÄÎä•
 
plt.figure(figsize=(8, 3))
for i in range(8):
    plt.subplot(1, 8, i + 1)

    # v[i]: Ïó¨Í∏∞ÏÑú v[i]Îäî Î≤°ÌÑ∞ÌôîÎêú Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞Ïùò iÎ≤àÏß∏ ÏïÑÏù¥ÌÖú
    # aspect: ÎßâÎåÄÍ∏∞ Í∞ÄÎ°úÍ∏∏Ïù¥
    # cmap=plt.cm.bone_r: Îßµ ÏÉâÏÉÅ
    plt.imshow(v[i], aspect=0.4,
               interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False); plt.xticks([]); plt.yticks([])
    plt.title("vector {}".format(i + 1))
plt.suptitle("vetorized images", y=1.05)
plt.tight_layout(w_pad=7)
plt.show()
```
