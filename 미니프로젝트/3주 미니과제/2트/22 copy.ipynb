## time_series_covid_19_confirmed.csv
> ## 신종 코로나바이러스 2019 데이터셋


```python
train_dataset = pd.read_csv('./time_series_covid_19_confirmed.csv')

# print(train_dataset.head)


drop_clo = ['Province/State','Country/Region','Lat','Long']
train_dataset=train_dataset.drop(drop_clo,axis=1)
datewise= list(train_dataset.columns)
val_dataset = train_dataset[datewise[-30:]]
print(train_dataset.head)
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# 데이터 분할 비율 설정
train_ratio = 0.8 # 훈련데이터 비율
num_dates = len(train_dataset.columns) - 4  # 2020/1월~ 2021/5월 -> 날짜 열의 개수 계산
num_train = int(num_dates * train_ratio) # 훈련데이터 개수
num_val = num_dates - num_train  # 검증데이터 개수

# 날짜 컬럼만 분할
date_columns = train_dataset.columns[4:]
train_dates = date_columns[:num_train]
val_dates = date_columns[num_train:]

# 시각화 생성
fig = make_subplots(rows=30, cols=1)

for i in range(30):  # 행별로 3개의 지역 시각화
    fig.add_trace(
        go.Scatter(x=np.arange(num_train), mode='lines+markers', y=train_dataset.loc[i, train_dates].values, marker=dict(color="dodgerblue"), showlegend=False),
        row=i+1, col=1
    )
    fig.add_trace(
        go.Scatter(x=np.arange(num_train, num_train + num_val), y=train_dataset.loc[i, val_dates].values, mode='lines+markers', marker=dict(color="darkorange"), showlegend=False),
        row=i+1, col=1
    )

fig.update_layout(height=1200, width=800, title_text="Train (blue) vs. Validation (orange) sales")
fig.show()

```
